# Get Youtube Channel Id from Youtube Channel url

````python
# Import the required modules
!pip install pandas
import requests
import pandas as pd
import re
# Get channel_id from channel_name of youtuber
def get_channelId(channel_link):

    # Get the HTML content of the channel page
    response = requests.get(channel_link).text

    # Find the canonical link tag in the HTML using a regular expression
    match = re.search(r'<link rel="canonical" href="https://www.youtube.com/channel/(.+?)"', response)

    # Check if a match was found
    if match:
        # Extract the channel ID from the match group
        channel_id = match.group(1)
        # Print the channel ID
        print(f"The channel ID for {channel_link} is {channel_id}")
        return channel_id
    else:
        # Print an error message
        print(f"No channel ID found for {channel_link}")
        return ''
````

# Iterate over all youtube videos for a specific channel

> Indented block



````python
import requests
import pandas as pd

def allVideos(channelurl):
    # Insert youtube channel video url here...
    channel_id = get_channelId(channelurl)
    print(channel_id)
    api_key = ""  # Replace with your own API key

    # Initialize an empty list to store DataFrames
    df_list = []

    # Make the first request to get the initial results and the total number of results
    url = f"https://youtube.googleapis.com/youtube/v3/search?part=snippet&channelId={channel_id}&maxResults=50&order=date&type=video&key={api_key}"
    response = requests.get(url=url).json()
    total_results = response["pageInfo"]["totalResults"]

    # Loop through the response items and append them to the list
    for i in response["items"]:
        channel_id = i["snippet"]["channelId"]
        video_id = 'https://www.youtube.com/watch?v=' + i["id"]["videoId"]
        video_title = i["snippet"]["title"]
        published_date = i["snippet"]["publishedAt"]
        vid_type = i["id"]["kind"]
        row_data = {"channel_id": channel_id, "video_id": video_id, "video_title": video_title, "published_date": published_date, "type": vid_type}
        df_list.append(pd.DataFrame([row_data]))

    # If there are more than 50 results, use the nextPageToken to get the next page of results
    if "nextPageToken" in response:
        page_token = response["nextPageToken"]
    else:
        page_token = None

    # Loop until there are no more pages of results
    while page_token is not None:
        # Make the next request with the page token
        url = f"https://youtube.googleapis.com/youtube/v3/search?part=snippet&channelId={channel_id}&maxResults=50&order=date&type=video&pageToken={page_token}&key={api_key}"
        response = requests.get(url=url).json()
        # Update the page token
        if "nextPageToken" in response:
            page_token = response["nextPageToken"]
        else:
            page_token = None
        # Loop through the response items and append them to the list
        for i in response["items"]:
            channel_id = i["snippet"]["channelId"]
            video_id = 'https://www.youtube.com/watch?v=' + i["id"]["videoId"]
            video_title = i["snippet"]["title"]
            published_date = i["snippet"]["publishedAt"]
            vid_type = i["id"]["kind"]
            row_data = {"channel_id": channel_id, "video_id": video_id, "video_title": video_title, "published_date": published_date, "type": vid_type}
            df_list.append(pd.DataFrame([row_data]))

    # Concatenate all DataFrames in the list into one DataFrame
    df = pd.concat(df_list, ignore_index=True)

    # Save the DataFrame as a CSV file
    df.to_csv("video_ids.csv", index=False)

# Define your get_channelId function here if it's not already defined
````

````python
import csv

# Specify the CSV file path
csv_file_path = 'video_ids.csv'

# Specify the column index you want to read (0-based index)
column_index = 1  # Change this to the index of the column you want to read

# Initialize an empty list to store the values from the column
column_values = []

# Open the CSV file and read the specified column with explicit encoding
with open(csv_file_path, mode='r', encoding='utf-8-sig') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        if len(row) > column_index:  # Check if the row has enough columns
            column_values.append(row[column_index])

# Now, column_values contains the values from the specified column as a list
print(column_values[1:])

````

# Transcribe the videos from youtuber example

````python
import os
import yt_dlp

os.environ['OPENAI_API_KEY'] = '';
urls = column_values[1:]#'https://www.youtube.com/watch?v=fh-y3tPyZ_I'
save_dir = "docs/youtube/"

loader = GenericLoader(
    YoutubeAudioLoader([urls], save_dir),
    OpenAIWhisperParser()
)
docs = loader.load()

````

# Summarize the transcript

````python
import openai
from openai import AsyncOpenAI
async def get_completion(prompt, model="gpt-3.5-turbo"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    client = AsyncOpenAI()
    completion = await client.chat.completions.create(model="gpt-4", messages=[{"role": "user", "content": prompt}])
    return completion.choices[0].message.content
````

# Summarize all YouTuber videos and save those in database or csv

````python
import csv
import openai
existing_data = []

for doc in docs:
    prompt = f"""
    Your task is to generate a short summary of a below-mentioned text \
    
    Summarize the transcript below, delimited by triple 
    backticks, in at most 30 words. 
    
    Review: ```{doc.page_content}```
    """
    video_summary = await get_completion(prompt)
    print(video_summary)
    source = (doc.metadata["source"]+'').replace('docs\\youtube\\','').replace('.m4a','')
    x = {"Video Name": source , "Transcription": doc.page_content, "Video Summary": video_summary}
    existing_data.append(x)

# Name of the CSV file
csv_file_name = "summarize_youtube_video.csv"

# Determine if the CSV file already exists
file_exists = True
try:
    with open(csv_file_name, mode='r') as file:
        file_exists = True
except FileNotFoundError:
    file_exists = False

# Write the data (including the new entry) back to the CSV file
with open(csv_file_name, mode='w', newline='', encoding='utf-8') as file:
    fieldnames = ["Video Name", "Transcription","Video Summary"]
    writer = csv.DictWriter(file, fieldnames=fieldnames, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    # Write the header only if the file is newly created
    if not file_exists:
        writer.writeheader()
    
    # Write the data to the CSV file
    writer.writerows(existing_data)

print(f"Data has been added to {csv_file_name}")
````

````python
# Split documents into chunks.
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1500,
    chunk_overlap = 150
)

splits = text_splitter.split_documents(docs)
````

````python
from langchain.vectorstores import Chroma
from langchain.embeddings.openai import OpenAIEmbeddings
embedding = OpenAIEmbeddings()

persist_directory = 'docs/chroma/'

vectordb = Chroma.from_documents(
    documents=splits,
    embedding=embedding,
    persist_directory=persist_directory
)

print(vectordb._collection.count())
````
